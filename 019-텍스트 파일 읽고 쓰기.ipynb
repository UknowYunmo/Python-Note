```python
텍스트 파일을 읽고 그 내용을 화면에 출력하고자 하면

제일 먼저 텍스트 읽기 모드로 파일을 연다.

텍스트 파일을 오픈하면 텍스트 파일을 읽어 내용을 화면에 출력할 수 있다.
```

```python
# 예제 : 스티브 잡스 연설문 불러와서 출력하기

f=open('c:\\data\\jobs.txt',encoding='UTF-8')
data=f.read()
print(data)
f.close()

"""
f.close()를 하지 않으면, 파이썬을 실행하고 있는 동안 파일이 계속 열려있게 된다.
-> 메모리를 계속 잡하먹힌다.
"""
```

```python
# 예제 : mydata3 라는 txt 파일을 읽고, 빅데이터 가 몇번 들어있는지 세기

f=open('c:\\data\\mydata3.txt',encoding='UTF-8')
data=f.read()
print(data.count('빅데이터'))
f.close()

"""
이렇게 짧은 텍스트는 한 번에 읽어와도 문제 없지만
텍스트 파일의 용량이 매우 클 경우 read()로 한꺼번에 읽어오는건
메모리 문제를 일으킬 수 있다.


웹에서 스크롤링한 데이터를 분석하고자 할 때 위와 같이 open 함수를 이용해서
파이썬으로 데이터를 불러온다.
"""
```

```python
# 예제 : 텍스트 파일을 한 줄씩 읽고 출력하기 ( readline )

"""
텍스트 파일의 용량이 매우 클 경우 read()로 읽지 말고
readline으로 한 줄씩 읽을 수 있다.
한 줄을 읽고 나면 파일을 읽기 시작하는 위치는 그 다음 줄의 맨 처음이 된다.
"""

f=open('c:\\data\\jobs.txt',encoding='UTF-8')
data=f.readline()
print(data)
f.close()

#첫째줄만 출력이 되었다.

#그렇다면 이제 loop문으로 전체를 읽어보자.

f=open('c:\\data\\jobs.txt',encoding='UTF-8')
data=f.readline()
while data:
    print(data)
    data=f.readline()
f.close()
```

```python
# 예제 : mydata3에서 readline을 활용해서 특정 단어가 몇 번 나왔는지 출력하기

f=open('c:\\data\\mydata3.txt',encoding='UTF-8')
data=f.readline()
cnt=0
find=input('검색할 단어를 입력하세요 : ')
while data:
    cnt+=data.count(find)
    data=f.readline()
f.close()
print(cnt)
```

```python
# 예제 : 내용을 적어 txt 파일 만들기

text=input('파일에 들어갈 내용을 입력하세요 : ')
f=open('c:\\data\\mydata.txt','w')
f.write(text)
f.close()
```

```python
# write()

listdata=[2,2,1,3,8,5,7]
result=sorted(listdata)
print(result)
f=open('c:\\data\\mydata2.txt',"w")
f.write(str(result))
f.close()
```

```python
# writelines()

data=[]
while True:
    text=input('저장할 내용을 입력하세요 : ')
    if text=="":
        break
    data.append(text+'\n')
f=open('c:\\data\\mydata99.txt','w')
f.writelines(data)
f.close()
```

```python
# 그냥 'w'로 동일한 파일위에 작성하면, 텍스트 내용이 싹 다 새로 덮어지지만,
# 이렇게 open에서 옵션을 'a'로 하면, 위에 내용을 추가할 수 있다.
data=[]
while True:
    text=input('저장할 내용을 입력하세요 : ')
    if text=="":
        break
    data.append(text+'\n')
f=open('c:\\data\\mydata99.txt','a')
f.writelines(data)
f.close()
```

```python
# 파이썬으로 텍스트 파일을 다른 이름으로 복사하기

f=open('c:\\data\\mydata.txt','r')
h=open('c:\\data\\mydata_copy.txt','w')
data=f.read()
h.write(data)
f.close()
h.close()
```

```python
# 바이너리 파일 (그림, 동영상, 노래, 실행 파일 등 )  복사하기

bufsize=1024 # 1kb 크기의 버퍼 사이즈
f=open('c:\\data\\lena.png','rb') # rb = read binary
h=open('c:\\data\\lena_copy.png','wb') # wb = write binary
data=f.read(bufsize) # 이미지를 1kb 읽어서 data에 저장
while data:
    h.write(data)
    data=f.read(bufsize)
f.close()
h.close()
```

```python
# 파일을 열고, 자동으로 닫기 ( with ~ as )

with open('c:\\data\\mydata.txt','r') as f:
    data=f.readlines()
    print(data)
    
# 이렇게 with ~ as 절을 사용하면 close()를 작성하지 않아도 자동으로 해준다.
```
